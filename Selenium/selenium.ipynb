```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
PATH = "C:\Program Files (x86)\chromedriver.exe"
```

```python
driver = webdriver.Chrome(executable_path=PATH)
```

```python
driver.get("https://ec.europa.eu/programmes/erasmus-plus/projects/eplus-project-details/#project/2019-1-TR01-KA201-074890")
```

```python
print(driver.page_source)
```

```python
name = driver.find_element_by_class_name("vlr-projectCard__coordinator_details_name")
address_street = driver.find_element_by_class_name("vlr-projectCard__coordinator_details_address_street")
address_city = driver.find_element_by_class_name("vlr-projectCard__coordinator_details_address_city")
address_region = driver.find_element_by_class_name("vlr-projectCard__coordinator_details_address_region")
web = driver.find_element_by_xpath("//div[@class='vlr-projectCard__coordinator_details_website']/a").text;
orgType = driver.find_element_by_class_name("vlr-projectCard__coordinator_details_orgType")

#coordinator_detail_name = driver.find_element_by_class_name("vlr-projectCard__coordinator_details_name")

print(name.text)
print(address_street.text)
print(address_city.text)
print(address_region.text)
print(web)
print(orgType.text)
```
